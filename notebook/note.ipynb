```python
import requests
```

```python
response = requests.get("https://docs.langchain.com/llms.txt")
text = response.text
text_lines = text.splitlines()
text_lines
```

```python
len(text_lines)
```

#### Extracting topics and urls

```python
import re
topics = re.findall(r"\[([^\[\]]+)\]", text)
topics
```

```python
urls = re.findall(r'https?://[^\s\]]+md',text)
urls
```

```python
from langchain_core.documents import Document
import uuid

document_structures=[]

for i,(topic,url) in enumerate(zip(topics,urls)):
    doc = Document(
        id=f"doc_{uuid.uuid4().hex[:8]}_{i}",
        page_content=topic,
        metadata={
            "url":url
        }
    )
    document_structures.append(doc)
```

```python
document_structures[1]
```

```python
import chromadb

client = chromadb.Client()
collection = client.get_or_create_collection("sample")

for doc in document_structures:
    collection.add(
        ids=doc.id,
        documents=doc.page_content,
        metadatas=doc.metadata
    )
```

```python
collection.count()
```

```python
collection.peek()
```

```python
query_result_topic = collection.query(query_texts="Langsmith collector proxy",n_results=3)
```

```python
query_result_topic
```

```python
retrieved_docs = []

if query_result_topic['documents'] and query_result_topic['documents'][0]:
    documents = query_result_topic['documents'][0]
    metadatas = query_result_topic['metadatas'][0]
    distances = query_result_topic['distances'][0]
    ids = query_result_topic['ids'][0]

    for i,(doc_id,document,metadata,distance) in enumerate(zip(ids,documents,metadatas,distances)):
        similarity_score = 1- distance

        if similarity_score >= 0.5:
            retrieved_docs.append({
                'doc_id':doc_id,
                'topic':document,
                'url':metadata.get('url'),
                'similarity_score':similarity_score,
                'distance':distance,
                'rank':i+1
            })

    print(f"Retrieved {len(retrieved_docs)} documents (after filtering)")
else: 
    print("No documents found")
```
